```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot, plot
import plotly.graph_objects as go
from datetime import datetime, timedelta
import sys, os
```

```python
df = pd.read_csv(os.path.join(os.path.dirname(os.path.dirname(os.path.abspath('file'))), 'data', 'pib_pais.csv'), sep = ';')
df
```

```python
df = df[123:205]
df
```

```python
df.reset_index(drop = True, inplace = True)
df
```

```python
#print(df.groupby(['Año', 'Periodo'])['Variación trimestral (en %)'].sum())
df["x"]=df['Periodo'].replace(regex = "rimestre ", value="") + '-' + df['Año'].str[2:]
df
```

```python
df['Variación trimestral (en %)'].replace(regex = ',', value = '.', inplace = True)
df
```

```python
df['Variación trimestral (en %)'] = df['Variación trimestral (en %)'].astype(float)
df
```

```python
maximo = np.ones(len(df)) * df['Variación trimestral (en %)'].max()
minimo = np.ones(len(df)) * df['Variación trimestral (en %)'].min()
```

```python
trace = go.Scatter(x = df['x'],
                  y = df['Variación trimestral (en %)'],
                  name = 'Variación trimestral del PIB (en %)',
                  mode= 'lines')

trace_h1 = go.Scatter(x = df['x'],
                  y = maximo,
                  name = 'Máximo Histórico',
                  mode= 'lines',
                  opacity = 0.4,
                  marker=dict(
                          color='green',
                          size=10,
                          ))

trace_h2 = go.Scatter(x = df['x'],
                  y = minimo,
                  name = 'Mínimo Histórico',
                  mode= 'lines',
                  opacity = 0.4,
                  marker=dict(
                          color='red',
                          size=10,
                          ))

data = [trace, trace_h1, trace_h2]

layout = dict(title = 'Variación trimestral del PIB (en %) (2001-2021)',
             xaxis= dict(title= 'Trimestres por año',ticklen= 5),
             yaxis= dict(title= '%',ticklen= 5)
           )

fig = go.Figure(data = data, layout=layout)

iplot(fig)

fig.write_html(os.path.join(os.path.dirname(os.path.dirname(os.path.abspath('file'))), 'data', 'pib.html'))
```
