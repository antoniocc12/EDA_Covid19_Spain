```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import init_notebook_mode, iplot, plot
import plotly.graph_objects as go
import sys, os
```

```python
df = pd.read_csv(os.path.join(os.path.dirname(os.path.dirname(os.path.abspath('file'))), 'data', 'vac_com.csv'), sep = ',', encoding = 'latin-1')
df
```

```python
df = df.iloc[:, 0:11]
df
```

```python
df = df.rename(columns={'Unnamed: 0':'Comunidad',
                        'Dosis entregadas Pfizer (1)':'Entregadas Pfizer',
                        'Dosis entregadas Moderna (1)':'Entregadas Moderna',
                        'Dosis entregadas AstraZeneca (1)':'Entregadas AstraZeneca',
                        'Dosis entregadas Janssen (1)':'Entregadas Janssen',
                        'Total Dosis entregadas (1)':'Total Entregadas',
                        'Dosis administradas (2)':'Administradas',
                        '% sobre entregadas':'% Administradas/Entregadas',
                        'Nº Personas con al menos 1 dosis':'Personas con 1 dosis',
                        'Nº Personas vacunadas\n(pauta completada)':'Personas con 2 dosis',
                        'Nº Personas con dosis adicional':'Personas con dosis adicional'})
df = df.iloc[0:22]
df
```

```python
# Exportamos para tableau

df.to_csv(os.path.join(os.path.dirname(os.path.dirname(os.path.abspath('file'))), 'data', 'vac_com_tableau.csv'), sep=',')
```

```python
nombres = list(df.columns[1:5])
val = list(df.iloc[21][1:5])
valores = []
for i in val:
    i = i.replace(',', '')
    valores.append(i)
print(valores, nombres)
```

```python
fig = go.Figure(data=[go.Pie(labels = nombres, values = valores, hole = 0.3, textfont_size = 15)], layout = go.Layout(height = 650,
                   width = 650,
                   autosize = False,
                   title = 'Tipos de Vacunas Entregadas',
                   title_font_size = 20,
                   legend_font_size = 15))
iplot(fig)

fig.write_html(os.path.join(os.path.dirname(os.path.dirname(os.path.abspath('file'))), 'data', 'pie_vac.html'))
```
